```python
# create a basic map in folium
map_osm=folium.Map(location=[-33.925, 18.625], zoom_start=10)
map_osm
```

```python
# add another basemap
map_dark=folium.Map(location=[-33.925, 18.625], tiles='CartoDB dark_matter', zoom_start=10)
map_dark
```

```python
# create a dataframe with needed data
df_suburbsloc=df_suburbs[['Latitude','Longitude']]
# df_suburbsloc.head()

# change df into a list
suburbs_list=df_suburbsloc.values.tolist()
suburbs_size=len(suburbs_list)  #to check the length of the suburbs, how many
suburbs_size
```

```python
# plotting the data on the map by iterating
for point in range (0, suburbs_size):
    folium.Marker(suburbs_list[point]).add_to(map_osm)
    folium.Marker(suburbs_list[point]).add_to(map_dark)

map_osm
```

```python
map_dark
```

```python
# adding pop-up message with name of suburb
for point in range (0, suburbs_size):
    folium.Marker(suburbs_list[point], popup=df_suburbs['Suburb'][point]).add_to(map_osm)
    folium.Marker(suburbs_list[point], popup=df_suburbs['Suburb'][point]).add_to(map_dark)

map_osm
```

```python
map_dark
```

```python
# change marker icon/styling
for point in range(0, suburbs_size):
    folium.Marker(suburbs_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='darkblue', icon_color='white',icon='tint', angle=0, prefix='fa')).add_to(map_osm)
    folium.Marker(suburbs_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='darkblue', icon_color='white',icon='tint', angle=0, prefix='fa')).add_to(map_dark)

map_osm
```

```python
map_dark
```

```python
# Classifying Markers
for point in range (0, suburbs_size):
    if df_suburbs['Oct 2017\nkl/month'][point]<5:
        folium.Marker(suburbs_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='green', icon_color='white',icon='thumbs-up', angle=0, prefix='fa')).add_to(map_osm)
        folium.Marker(suburbs_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='green', icon_color='white',icon='thumbs-up', angle=0, prefix='fa')).add_to(map_dark)
    elif df_suburbs['Oct 2017\nkl/month'][point]>5 and df_suburbs['Oct 2017\nkl/month'][point]<=8:
        folium.Marker(suburbs_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='blue', icon_color='white',icon='thumbs-up', angle=0, prefix='fa')).add_to(map_osm)
        folium.Marker(suburbs_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='blue', icon_color='white',icon='thumbs-up', angle=0, prefix='fa')).add_to(map_dark)
    else:
        folium.Marker(suburbs_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='red', icon_color='white',icon='thumbs-down', angle=0, prefix='fa')).add_to(map_osm)
        folium.Marker(suburbs_list[point], popup=df_suburbs['Suburb'][point], icon=folium.Icon(color='red', icon_color='white',icon='thumbs-down', angle=0, prefix='fa')).add_to(map_dark)
map_osm      
        
```

```python
map_dark
```

```python
#creating a point map
map_points=folium.Map(location=[-33.35, 18.625], tiles='CartoDB dark_matter', zoom_start=10)
for point in range (0, suburbs_size):
    folium.Circle(
        location=suburbs_list[point],
        popup=df_suburbs['Suburb'][point]+ ':'+ str(df_suburbs['Oct 2017\nkl/month'][point])+' kL',
        radius=str(df_suburbs['Oct 2017\nkl/month'][point]* 100),
        color='#17cbef',
        fill=True,
        opacity=0.8,
        fill_color='#17cbef',
        stroke=True,
        weight=1.0,
        ).add_to(map_points)

map_points
        
```

```python
#Classifying Points
map_points=folium.Map(location=[-33.925,18.625], tiles='CartoDB dark_matter', zoom_start=10)
for point in range(0, suburbs_size):
    if df_suburbs['Oct 2017\nkl/month'][point]<=8 and df_suburbs['Oct 2017\nkl/month'][point]>0:
        usage_color='green'
    elif df_suburbs['Oct 2017\nkl/month'][point]>8 and df_suburbs['Oct 2017\nkl/month'][point]<=10:
        usage_color='#17cbef'
    elif df_suburbs['Oct 2017\nkl/month'][point]>10:
        usage_color='red'
    else: # Outliers
        usage_color='orange'

    folium.Circle(
        location=suburbs_list[point],
        popup=df_suburbs['Suburb'][point]+ ':'+ str(df_suburbs['Oct 2017\nkl/month'][point])+' kL',
        radius=str(df_suburbs['Oct 2017\nkl/month'][point]*100),
        color=usage_color,
        fill=True,
        opacity=0.8,
        fill_color=usage_color,
        stroke=True,
        weight=1.0        
    ).add_to(map_points)

map_points
```

```python
#Creating Choropleth maps
df_world_countries=pd.read_json('C:/Users/HomePC/Documents/GIS WORK/GIS PROJECTS/FOLIUM/world-countries.json')
df_unemployment=pd.read_csv('C:/Users/HomePC/Documents/GIS WORK/GIS PROJECTS/FOLIUM/unemployment.csv', skiprows=4,delimiter=',')

df_unemployment

```

```python
#cleaning the data
df_unemployment_2018=df_unemployment[['Country Name','Country Code','2018' ]]
df_unemployment_2018
```

```python
#remove Nans
df_unemployment_2018_clean=df_unemployment_2018.dropna()
df_unemployment_2018_clean
```

```python
#prepare data for map 
legend_title='Unemployment Rate Per Country 2018 (%)'

# filter dataframe to country code
data_unemployment=df_unemployment_2018_clean[['Country Code', '2018']]
data_unemployment

```

```python
#setting the map
map_unemployment=folium.Map(location=[8.7832, 34.5085], zoom_start=3)
folium.Choropleth(geo_data='C:/Users/HomePC/Documents/GIS WORK/GIS PROJECTS/FOLIUM/world-countries.json',
                            data=data_unemployment,
                            columns=['Country Code', '2018'],
                            key_on='feature.id',
                            fill_color='YlOrRd',
                            fill_opacity=0.7,
                            line_opacity=0.2,
                            legend_name=legend_title
                           ).add_to(map_unemployment)
map_unemployment
```

```python
#Measuring Boundaries
df_dam_levels=pd.read_csv('https://raw.githubusercontent.com/EBISYS/Mapping-Dam-Levels/refs/heads/master/DamLevels.txt', delimiter=',')
df_dam_levels
```

```python
df_location=pd.read_csv('https://raw.githubusercontent.com/EBISYS/Mapping-Dam-Levels/refs/heads/master/damLocations.txt', delimiter=',')
df_location
```

```python
#join the dams together on the dams field
df_dam_data=pd.merge(df_dam_levels, df_location, on='Dam', how='left')
df_dam_data
```

```python
#prepare plotting data
df_dams_locations=df_dam_data[['Latitude','Longitude']]

# convert into a list
dams_locations_list=df_dams_locations.values.tolist()
dams_locations_size=len(dams_locations_list)
dams_locations_size
```

```python
map_dams=folium.Map(location=[-33.925, 18.625], zoom_start=10)
for point in range (0, dams_locations_size):
    folium.Marker(dams_locations_list[point], popup=df_dam_data['Dam'][point]
    ).add_to(map_dams)
map_dams
```

```python
map_url=folium.Map(location=[-33.625, 18.625], 
                   tiles='https://services.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
                    zoom_start=12,
                   attr='ArcGIS Online'
                  )

for point in range (0, dams_locations_size):
    folium.Marker(dams_locations_list[point], popup=df_dam_data['Dam'][point]
    ).add_to(map_url)
map_url
```

```python
#measure control
from folium.plugins import MeasureControl

map_url.add_child(MeasureControl())
map_url
```

```python
#Creating Layers Control and Basemaps
map_layers=map_points

add='/MapServer/tile/{z}/{y}/{x}'
ESRI= dict(
    World_Ocean_Base='https://services.arcgisonline.com/ArcGIS/rest/services/World_Ocean_Base',
    World_Navigation_Charts='https://services.arcgisonline.com/ArcGIS/rest/services/Specialty/World_Navigation_Charts',
    NatGeo_World_Map='https://services.arcgisonline.com/ArcGIS/rest/services/NatGeo_World_Map/MapServer',
    World_Imagery='https://services.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer',
    World_Physical_Map='https://services.arcgisonline.com/ArcGIS/rest/services/World_Physical_Map/MapServer',
    World_Shaded_Relief='https://services.arcgisonline.com/ArcGIS/rest/services/World_Shaded_Relief/MapServer',
    World_Street_Map='https://services.arcgisonline.com/ArcGIS/rest/services/World_Street_Map/MapServer',
    World_Terrain_Base='https://services.arcgisonline.com/ArcGIS/rest/services/World_Terrain_Base/MapServer',
    World_Topo_map='https://services.arcgisonline.com/ArcGIS/rest/services/World_Topo_Map/MapServer'   
)

for tile_name, tile_url in ESRI.items():
    tile_url+=add
    folium.TileLayer(tile_url,
                    name=tile_name,
                    attr='ESRI').add_to(map_points)

folium.LayerControl().add_to(map_layers)
map_layers
```

```python
#Creating a heatmap
from folium.plugins import HeatMap

# plot points on the heat map
map_heat=folium.Map(location=[-33.925, 18.625], tile='CartoDB dark_matter',
                   zoom_start=10)
for point in range (0, suburbs_size):
    folium.Circle(
        location=suburbs_list[point],
        popup=df_suburbs['Suburb'][point] + ':' + str(df_suburbs['Oct 2017\nkl/month'][point])+'kL',
        radius='15',
        color='darkblue',
        fill=True,
        opacity=0.8,
        fill_color='darkblue',
        stroke=True,
        weight=1.0).add_to(map_heat)
                                                      
# create an array
suburbsArray=df_suburbs[['Latitude','Longitude']].values

# plot on the heat map
map_heat.add_child(HeatMap(suburbsArray))
map_heat
```
